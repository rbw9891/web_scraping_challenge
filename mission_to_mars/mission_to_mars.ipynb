```python
# dependencies
from splinter import Browser
from bs4 import BeautifulSoup as bs
from webdriver_manager.chrome import ChromeDriverManager
```

```python
# Setup splinter
# location of driver
executable_path = {'executable_path': ChromeDriverManager().install()}
# instantiate browser object, will open a new empty browser window
browser = Browser('chrome', **executable_path, headless=False)
```

```python
# set url to mars news site
url = "https://redplanetscience.com/"

# visit webpage with splinter
browser.visit(url)
```

```python
# have splinter grab the html
# create beautiful soup object from html
# check to see if splinter retrieves html that was missing from requests get
html = browser.html
soup = bs(html, "html.parser")
print(soup.prettify())
```

```python
# find the <section class="image_and_description_container" and
# assign the latest news title and paragraph to variables
# print to make sure correct
for results in soup.find_all("section", class_="image_and_description_container"):
    title = results.find("div", class_="content_title").text
    para = results.find("div", class_="article_teaser_body").text
    print(title)
    print(para)
```

```python
# quit browser
browser.quit()
```
