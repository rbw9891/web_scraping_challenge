```python
from bs4 import BeautifulSoup as bs
import requests
```

```python
# set url to mars news site
url = "https://redplanetscience.com/"

# set html to requests of url
html = requests.get(url)
```

```python
# check for status code
html.status_code
```

```python
# create soup object
soup = bs(html.text, 'html.parser')
```

```python
soup
```

```python
# look for <div class="content_title"
title_result = soup.body.find("section", class_="image_and_description_container")

for results in title_result:
    title = results.find_all("div", class_="content_title")
    print(title)
```
